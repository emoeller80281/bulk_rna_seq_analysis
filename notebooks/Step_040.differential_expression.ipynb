```python
import pandas as pd
import scanpy as sc
import anndata as ad
from pathlib import Path

figdir = Path("/gpfs/Home/esm5360/bulk_rna_seq_analysis/figures/differential_expression")
figdir.mkdir(parents=True, exist_ok=True)
sc.settings.figdir = str(figdir)
```

### Load combined WT and KO adata from Step_030

```python
adata = ad.read_h5ad("/gpfs/Home/esm5360/bulk_rna_seq_analysis/data/combined_wt_ko_adata.h5ad")
```

### Run differential expression using the Wilcoxon rank sum test on the leiden clusters

```python
# Differential expression between conditions
sc.tl.rank_genes_groups(
    adata,
    groupby="leiden",
    key_added="wilcoxon",
    method="wilcoxon",
    use_raw=False,
    pts=True,
)
```

### Run hierarchical clustering using the PCA distances for the leiden clusters

```python
sc.tl.dendrogram(adata, groupby="leiden", use_rep="X_pca", key_added="leiden_dendrogram")
```

### Plot differentially expressed genes

1 cluster vs all other clusters

```python
sc.pl.rank_genes_groups_heatmap(adata, n_genes=10, key="wilcoxon", groupby="leiden", save="clusters_diff_expr.png")
sc.pl.rank_genes_groups_dotplot(adata, n_genes=10, key="wilcoxon", groupby="leiden", save="clusters_diff_expr.png")
```

### Save the differential expression results as a csv file

```python
de_table = sc.get.rank_genes_groups_df(adata, group=None, key="wilcoxon")
de_table.to_csv(
    "/gpfs/Home/esm5360/bulk_rna_seq_analysis/results/condition_de_wilcoxon.csv",
    index=False,
)

de_table.head()
```
